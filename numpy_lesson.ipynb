```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

#日本語フォント設定
mpl.rcParams['font.family'] = 'IPAexGothic'
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r--', label='比例')
plt.plot(t,t**2,'bs', label='平方')
plt.plot(t,t**3,'g^', label='立方')
plt.legend()
```
