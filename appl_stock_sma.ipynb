```python
import pandas_datareader
import matplotlib.pyplot as plt
import pandas as pd

import yfinance

yfinance.pdr_override()
```

```python
# Appleの株価データを取得
df_aapl = pandas_datareader.data.DataReader('AAPL', '2019-01-01')
```

```python
df_aapl['SMA'] = df_aapl['Close'].rolling(window=14).mean()
df_aapl['Close'].plot(label='Close', color = 'blue')
df_aapl['SMA'].plot(label='SMA', color = 'red')

#y軸を140から200の範囲で表示
plt.ylim([140, 200])

#x軸を2023-01-01から2024-06-01の範囲で表示
plt.xlim(['2023-01-01', '2024-06-01'])

plt.legend()

plt.show()
```
